```python
#import librarys to graphic
import numpy as np
import matplotlib.pyplot as plt

#creating an fucnction to create elements to 1 - 10 with size of 100 A
def create_data(n):
   return np.random.randint(10, size = n)

A = create_data(100)
print(A)


```

```python
import time

#Basic
def simple(A):
  result = A * 1

#Lineal
def linear(A):
  for element in A:
    element  = element * 2

#Cuadratic
def cuadratic(A):
  for i in A:
    for j in A:
      resu = i * j

#Cubic
def cubic(A):
  for k in A:
    for l in A:
      for m in A:
        res = k * l * m

#cubic sum double n
def plus_d(A):
  r = 0
  for a in A:
    for b in A:
      for c in A:
        r = a * b * c
        r += len(A) * 2


start = time.time()
linear(A)
cuadratic(A)
cubic(A)
plus_d(A)
simple(A)
end = time.time()
print("Time: ", end - start)
```

```python
import time

sizes = [10, 100, 1000]
linear_times = [ ]
cuadratic_times = [ ]
cubic_times = [ ]
plus_d_times = [ ]
simple_times = [ ]

#Basic/Simple way
for n in sizes:
    A = create_data(n)

    start = time.time()
    simple(A)
    end = time.time()

    total_time = end - start
    simple_times.append(total_time)

    print("N(simple):", n,"with time:", total_time)
#Linear
for n in sizes:
    A = create_data(n)

    start = time.time()
    linear(A)
    end = time.time()

    total_time = end - start
    linear_times.append(total_time)

    print("N(linear):", n, "with time:", total_time)
#Cuadratic
for n in sizes:
    A = create_data(n)

    start = time.time()
    cuadratic(A)
    end = time.time()

    total_time = end - start
    cuadratic_times.append(total_time)

    print("N(cuadratic)", n, "with time:", total_time)
#Cubic
for n in sizes:
    A = create_data(n)

    start = time.time()
    cubic(A)
    end = time.time()

    total_time = end - start
    cubic_times.append(total_time)

    print("N(cubic):", n, "with time:", total_time)
#Cubic sum double n
for n in sizes:
    A = create_data(n)

    start = time.time()
    plus_d(A)
    end = time.time()

    total_time = end - start
    plus_d_times.append(total_time)
    print("N(cubic plus double n):", n, "with time:",total_time)

```

```python
plt.scatter(sizes, simple_times)
plt.plot(sizes, simple_times, c = "m")
plt.scatter(sizes, linear_times)
plt.plot(sizes, linear_times, c = "r")
plt.scatter(sizes, cuadratic_times)
plt.plot(sizes, cuadratic_times, c = "b")
plt.scatter(sizes, cubic_times)
plt.plot(sizes, cubic_times, c = "y")
plt.scatter(sizes, plus_d_times)
plt.plot(sizes, plus_d_times, c = "g")
```
